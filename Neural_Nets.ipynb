```python
df.drop(['Unnamed: 32','id'],inplace=True,axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

```python
X_train=torch.from_numpy(X_train)
X_test=torch.from_numpy(X_test)
y_train=torch.from_numpy(y_train)
y_test=torch.from_numpy(y_test)
```

```python
X_train.shape
```

```python
class MyModel():
  def __init__(self,X) :
    self.weights=torch.rand(X.shape[1],dtype=torch.float64,requires_grad=True)
    self.bias=torch.zeros(1,dtype=torch.float64,requires_grad=True)

  def forward(self,X_train):
    z=torch.matmul(X_train,self.weights)+self.bias
    return torch.sigmoid(z)

  def loss(self,y_test,y_pred):
    y_pred = torch.clamp(y_pred, min=1e-7, max=1 - 1e-7)
    return (-y_test*torch.log(y_pred)-(1-y_test)*torch.log(1-y_pred)).mean()

  def backprop(self,loss):
    loss.backward()
    with torch.no_grad():
      self.weights-=0.01*self.weights.grad
      self.bias-=0.01*self.bias.grad
      self.weights.grad.zero_()
      self.bias.grad.zero_()
```

```python
model=MyModel(X_train)
```

```python
for i in range(100):
  y_pred=model.forward(X_train)
  loss=model.loss(y_train,y_pred)
  model.backprop(loss)
  print(f"Epochs : {i+1}  Loss : {loss}")
```

# **Using NN Module**

```python
import torch.nn as nn
from torchinfo import summary
```

```python
class Model(nn.Module):
  def __init__(self,features):

    super().__init__()
    self.h1=nn.Linear(features,3)
    self.relu=nn.ReLU()
    self.o1=nn.Linear(3,1)
    self.s1=nn.Sigmoid()

  def forward(self,features):
    layer1=self.h1(features)
    output=self.relu(layer1)
    layer2=self.o1(output)
    return self.s1(layer2)
```

```python
features=torch.rand(10,5)
```

```python
model=Model(features.shape[1])
model(features)
```

```python
new_model=nn.Sequential(nn.Linear(X_train.shape[1],1),nn.Sigmoid())
optimizer=torch.optim.SGD(new_model.parameters(),lr=0.01)
criterion=nn.BCELoss()
for i in range(10):
  new_model.train()

  y_pred=new_model(X_train.float())

  calculated_loss=criterion(y_pred,y_train.float().reshape(-1, 1))

  optimizer.zero_grad()

  calculated_loss.backward()

  optimizer.step()

  print(f"Epochs : {i+1}  Loss : {calculated_loss}")
```
